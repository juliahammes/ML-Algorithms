# <font color='green'>AdaBoost (Adaptive Boosting)</font>

AdaBoost é um algoritmo de Machine Learning usado como um método ensemble, ou seja, combina vários algoritmos básicos para formar um algoritmo preditivo otimizado.

Como o próprio nome já diz, o tipo de ensemble utilizado pelo algoritmo AdaBoost é o Boosting, ele tenta construir um aprendiz forte (modelo preditivo) a partir dos erros de vários modelos mais fracos, basicamente tenta reduzir o erro de viés que surge quando os modelos não são capazes de identificar tendências relevantes nos dados.

O algoritmo mais comum utilizado com Adaboost é árvore de decisão porém com apenas um nível.

## Passo a Passo do Algoritmo

#### Passo 01: Inicialização dos pesos das amostras

O primeiro passo do algoritmo AdaBoost é atribuir pesos a cada amostra de dados, os pesos de cada amostra indicam o quão importante é ser classificado corretamente, assim cada novo aprendiz irá se concentrar nessas amostras e irá tentar classificá-las corretamente. 

Inicialmente os pesos serão iguais para todas as amostras.

#### Passo 02: Construir uma árvore de decisão para cada feature e avaliar os resultados.

O algoritmo cria uma árvore de decisão, com apenas um nível, para cada uma das features dos dataset e então calcula o Index Gini de cada uma das árvores. A árvore com menor Index Gini será o primeiro aprendiz.

#### Passo 04: Atualizar os pesos das amostras

As amostras que o aprediz anterior classificou incorretamente receberão mais peso, enquanto os pesos das amostras com previsões corretas serão diminuídos.
Assim o próximo aprendiz dará preferência as amostras com maior peso.


adaboost3.png

O amount of say (alpha) será **negativo** quando a amostra é classificada corretamente, e será **positivo** quando a amostra é classificada incorretamente.


A soma total dos pesos das amostras deve ser 1, para isso será necessário normalizar os pesos (dividindo cada peso pela soma de todos os pesos).


#### Passo 05: Montar um novo dataset

A ideia aqui é criar um novo dataset que contenha cópias duplicadas das amostras com maiores pesos.

Para gerarmos esse novo dataset um número entre 0 e 1 é escolhido, e vemos aonde esse número cai quando usamos os pesos das amostras como uma distribuição. Então a amostra é selecionada para compor nosso dataset. Repetimos esse processo até compor um dataset de tamanho igual ao original.


#### Passo 06: Repetir os passos acima para o novo dataset

Agora desconsideramos o dataset anterior e com o novo dataset gerado iremos atribuir pesos iguais para todas as amostras. Como as amostras classificadas de forma incorreta anteriormente estão duplicadas nesse novo dataset, elas serão tratadas como um bloco criando uma penalidade grande por serem classificadas de maneira incorreta. 

Encontraremos a árvore que faz o melhor trabalho classificando a nova coleção de amostras encontrando seu Índice de Gini e selecionando aquele com o índice de Gini mais baixo.

Calculando o "Amount of Say" e o "Erro Total" para atualizar o pesos de amostras anteriores. 

Normalizando os novos pesos de amostra. 

Repitindo essas etapas até e a menos que um erro de treinamento baixo seja alcançado.

#### Gerando o resultado

Como vimos o AdaBoost será composto por uma floresta de árvores de decisão com apenas um nível, cada árvore de decisão irá gerar uma classificação para a nova observação, sendo assim como o algortimo define a qual classe pertence essa nova observação?

Para isso leva-se em consideração o "Amount of Say" de cada árvore. Soma-se o "Amount of say" das árvores (stumps) com a mesma classificação. Então a observação é classificada de acordo com a classe que teve a maior somatória de "Amount f say".
